```python
import requests
from bs4 import BeautifulSoup
import time
import datetime

import numpy as np
import pandas as pd
```

```python
URL_uk_cities = "https://www.timeanddate.com/weather/uk"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page_uk_cities = requests.get(URL_uk_cities, headers=headers)

soup11 = BeautifulSoup(page_uk_cities.content, "html.parser")

df_weather = pd.DataFrame()
for city in soup11.find_all('a', href=True):
    if '/weather/uk/' in city['href']:
        uk_city = city.text.replace(" ", "-")
        #print(uk_city)

        URL_cities_weather = f"https://www.timeanddate.com/weather/uk/{uk_city}/ext"
        lic = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

        page_cities_weather = requests.get(URL_cities_weather, headers=lic)
        
        if page_cities_weather:
            soup21 = BeautifulSoup(page_cities_weather.content, "html.parser")
        
            headers = [th.text for th in soup21.select('thead th') if th.text != '&nbsp;' and th.text != '\xa0']
            headers = headers[4:]

            values = [td.text for td in soup21.select('tbody td') if td.text != '' and td.text!='↑']

            days = [datetime.date.today() + datetime.timedelta(days=i) for i in range(15)]

            data_dict = {}

            for i in range(0, len(headers)):
                if headers[i] == 'Day':
                    data_dict[headers[i]] = days
                else:
                    data_dict[headers[i]] = values[i-1::10]
            
            data_dict['City'] = [uk_city] * 15
            
            df = pd.DataFrame(data_dict)

            df_weather = pd.concat([df_weather, df], ignore_index=True)
```
